```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('20251231_111204_commands.csv')
data = df.values.astype(float)

# Display the first few rows
time = data[:, 0]
# commands = data[:,1:8]
# vx = data[:,1]
# vy = data[:,2]
# wz = data[:,3]
# bcs_x = data[:,4]
# bcs_y = data[:,5]
# wz_nav = data[:,6]
# mode = data[:,7]

# phase = data[:,1:3]
# commands = data[:,3:8]

commands = data[:, 1:4]
nav_commands = data[:, 4:7]
phase = data[:, 7:9]
mode = data[:, 9]
obs_commands = data[:, 10:15]
```

```python
time_start = 0
time_end = -1

time_start = 60 * 100
time_end = 80 * 100

plt.plot(time[time_start:time_end], nav_commands[time_start:time_end], label=["bcs_x", "bcs_y", "wz_nav"])
plt.grid()
plt.legend()
```

```python
dist = np.sqrt(nav_commands[:, 0]**2 + nav_commands[:, 1]**2)
thres = np.ones_like(dist) * 0.1
plt.plot(time[time_start:time_end], dist[time_start:time_end], label="dist")
plt.plot(time[time_start:time_end], thres[time_start:time_end], label="thres")
plt.grid()
plt.legend()
```

```python
thres = np.ones_like(dist) * 0.2
plt.plot(time[time_start:time_end], np.abs(nav_commands[time_start:time_end, 2]), label="angle")
plt.plot(time[time_start:time_end], thres[time_start:time_end], label="thres")
plt.grid()
plt.legend()
```
